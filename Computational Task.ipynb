```python
import random
import string
import pandas as pd
import matplotlib.pyplot as plt

N=20 #Number of offsprings
master_sequence = 'unlearning is just as important as learning.'
l = len(master_sequence) #44
p= 0.025 #Probability of a mutation

def initialGens():
    gen = ''.join(random.choice(string.ascii_letters + string.punctuation + " ") for x in range(len(master_sequence)))  
    #Generates random string the same length of master sequence.
    return (str(gen))

def hamming (str1, str2):
    if not len(str1) == len(str2): #Checks if strings are of the same length
        print('error')
        return -1
    
    else:
        distance = 0
        
        for i in range(len(str1)):
            if not (str1[i] == str2[i]):
                distance = distance + 1 #Calculates hamming distance between two strings of the same length
                
        return (distance)

def mutate(i,chars = string.ascii_letters + string.punctuation +" "):
    x = random.random()
    
    if (x<p):
        return (random.choice(chars)) #Determines whether a mutation occurs. p probability of happening
    else:
        return i #returns original letter if no mutation occurs
    
def offspring(offsprings):
    for i in offsprings:
        distances.append(hamming(i, master_sequence))

    df = pd.DataFrame(list(zip(distances,offsprings)), columns=['distance', 'gen']) 
    #df['distance'].idxmin()
    gen = df.iloc[df['distance'].idxmin()]['gen'] #Selects the offspring with the smallest hamming distance
    return (gen)
```

```python
gens = []
distances = []
for i in range(N):
    gens.append(initialGens())
```

```python
gen = offspring(gens)
gen
```

```python
#gen='unlearning is just as important as learninp.'
```

```python
gen=gen.lower()
distance = hamming(master_sequence, gen)
count = 0
hDist = []
genNum = []

print("Generation ", count,": ", gen, l-distance)


while distance !=0:
    offsprings = []
    count = count + 1
    offsprings.append(gen)
    gen=list(gen)
    
    for j in range(N):
        newGen=[]
        for i in range(l):
    #         if not gen[i]==master_sequence[i]: #checks if char matches master sequence
                newGen.append(mutate(gen[i])) #mutation
        newGen = "".join(newGen)
        offsprings.append(newGen)
        gen = "".join(gen)
    newGen = offspring(offsprings)

    distance = hamming(master_sequence, gen)
    newDistance = hamming(master_sequence, newGen)
    
    
    if(newDistance==0):
        print('----------Master Sequence Has been Acheived-----------')
        print('Probability: ', p)
        print('Number of offspring: ',N)
        
        distance=newDistance
        gen=newGen
    elif(newDistance<distance):
        gen=newGen
        distance=newDistance
    elif(newDistance>distance):
        gen=gen
    
    correctChars = l-distance
    if(count%1000 ==0 or correctChars == l):
        print('Generation', count, ': ', gen, correctChars)
        hDist.append(correctChars)
        genNum.append(count)

    
        
```

Given the data I have collected above, it is clear to me that the magnitude of changing the probability is larger than that of the number of offsprings, when it comes to reaching the master sequence.

Probability (p):

There is an optimum point around about the 0.025 region as shown by the results. In experiments 2 and 4, the number of offsprings were kept the same but p were 0.01 and 0.05, it took these an average of 54972 and 40899 respectively across 8 trials, whereas in experiment 12 where p = 0.25, the average was significantly lower at 28663. This leads me to think that the probability is a normal distribution where the mean is somewhere between 0.05 and 0.01. This is further proven by the fact that when p < 0.01 and p > 0.05 the number of generations increase until a point where it takes too many generations to come to the master sequence.

Number of offsprings (N):

My initial thoughts were that the more offsprings made in each generation, the lower the number of generations needed to reach the master sequence. However this did not seem to be the case. Experiments 4, 5 and 6 shows a slight increase in the number of generations taken despite the the number of offsprings being raised from 20 to 50 to 150. There is either a very weak positive correlation or no correlation at all between the number of offsprings and the number of generations needed to reach the master sequence.

These trends described are not conclusive as only 7 trials for each case were completed.
